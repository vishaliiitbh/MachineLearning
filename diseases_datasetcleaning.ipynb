```python
t=data.drop(["Age","Gender"],axis=1)
t
```

```python
t.dropna()
```

```python
b=pd.get_dummies(t['Cough'])
b.rename(columns = {'Yes':'Cough'}, inplace = True)
b=b.drop(['No'],axis=1)

c=pd.get_dummies(t['Fatigue'])
c.rename(columns = {'Yes':'Fatigue'}, inplace = True)
c=c.drop(['No'],axis=1)

d=pd.get_dummies(t['Difficulty Breathing'])
d.rename(columns = {'Yes':'Difficulty Breathing'}, inplace = True)
d=d.drop(['No'],axis=1)

e=pd.get_dummies(t['Outcome Variable'])
e.rename(columns = {'Positive':'Outcome Variable'}, inplace = True)
e=e.drop(['Negative'],axis=1)

a=pd.get_dummies(t['Fever'])
a.rename(columns = {'Yes':'Fever'}, inplace = True)
a=a.drop(['No'],axis=1)
```

```python
t=pd.concat([a,b,c,d,e],axis=1)
```
